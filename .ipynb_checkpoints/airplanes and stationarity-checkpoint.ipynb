A time series is said considered stationnary if:
1. constant mean
2. constant variance
3. autocovariance that does not depend on time



```python
#plotting a time series is easy
ts.plot()
```

Obviously an increasing trend with cyclicality
The following function wil test for stationarity by checking the dickey fuller statistics and plotting a rolling std.

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean =timeseries.rolling(window=12,center=False).mean()
    rolstd = timeseries.rolling(window=12,center=False).std()
    #rolmean = pd.rolling_mean(timeseries, window=12)
    #rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
test_stationarity(ts)
```

Making the time series STATIONARY

2 things prevent a series from being stationary. That's the trend and seasonality.


Estimating & Eliminating Trend

quickest way to retunce trend can be transformation. 
For example let's start with something that will penalize higher values more than smaller values.
Let's start with a log transform.

```python
ts_log=np.log(ts)

ts_log.plot()
```

To remove the TREND:

Aggregation – taking average for a time period like monthly/weekly averages
Smoothing – taking rolling averages
Polynomial Fitting – fit a regression model

In this case we'll be doing some SMOOTHING, 

```python
moving_avg=ts_log.rolling(window=12).mean()

plt.plot(ts_log)
plt.plot(moving_avg,color='red')

```

Now we'll be looking to subtract the mean from the oscillations

```python
ts_log_moving_avg_diff=ts_log-moving_avg
ts_log_moving_avg_diff.head(12)
```

```python
#because your average needs 12 months before it can be calculated
#the first 12 data points of results are going to be missing. 

ts_log_moving_avg_diff.dropna(inplace=True)
test_stationarity(ts_log_moving_avg_diff)

```

Notice that if the test statistic is smaller than the 5% critical value, it means that we can say with 95% confidence that this is a stationary series.

DRAWBACK: we can take yearly averages, but in complex situations like forecasting a stock price, its difficult to come up with numbers.

Many use an MA that puts more emphasis on the recent prices. EXPONENTIAL MOVING AVERAGE

```python
expwighted_avg = ts_log.ewm( halflife=12).mean()
#halflife is the amount of exponential decay, this of course can vary
#typically it varies with the business domain it describes.

plt.plot(ts_log)
plt.plot(expwighted_avg, color='red')
```

Removing the EMA...

```python
ts_log_ewma_diff = ts_log - expwighted_avg
test_stationarity(ts_log_ewma_diff)
```

Notice that this one has an even less variations in mean and std. Also you can see that the test statistic is lower than the 1% critical value, meaning there's over a 99% chance that it's now a time series. 

ELIMINATING SEASONALITY

Here we'll deal with techniques for removing both trends and seasonality

Differencing – taking the differece with a particular time lag
Decomposition – modeling both trend and seasonality and removing them from the model.



```python
#first order differencing can be done with pandas
ts_log_diff = ts_log - ts_log.shift()  #think of shift as just 
#shifting the column by 1 value. Everything is off by 1 time period
#it's 1 period by default, but you can set other time periods.
plt.plot(ts_log_diff)
```

```python
ts_log_diff.dropna(inplace=True)
test_stationarity(ts_log_diff)
```

YOU ARE FREE TO REPEAT THIS PROCESS A FEW MORE TIMES, MIGHT IMPROVE YOUR RESULTS.



DECOMPOSING
both trend and seasonality are modeled separately and the remaining part of the series is returned

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts_log)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(ts_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

The function took out both the trend and the seasonality, we are left with the residulas, which we take. 

```python
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)
test_stationarity(ts_log_decompose)
```

Again, notice how our test statistic is lower than the 1% critical value, by a lot.

Keep in mind though, that converting these BACK into their original prices, is not very intuitive.

FORECASTING A TIME SERIES

Once you've taken out the trend and cyclicality, there can be two situations.

There can be 2 situations:

1. A strictly stationary series with no dependence among the values. This is the easy case wherein we can model the residuals as white noise. But this is a very rare. 

2.A series with significant dependence among values. In this case we need to use some statistical models like ARIMA to forecast the data. 

The ARIMA forecasting for a stationary time series is nothing but a linear (like a linear regression) equation. The predictors depend on the parameters (p,d,q) of the ARIMA model:

1.Number of AR (Auto-Regressive) terms (p): AR terms are just lags of dependent variable. For instance if p is 5, the predictors for x(t) will be x(t-1)….x(t-5).

2.Number of MA (Moving Average) terms (q): MA terms are lagged forecast errors in prediction equation. For instance if q is 5, the predictors for x(t) will be e(t-1)….e(t-5) where e(i) is the difference between the moving average at ith instant and actual value.

3.Number of Differences (d): These are the number of nonseasonal differences, i.e. in this case we took the first order difference. So either we can pass that variable and put d=0 or pass the original variable and put d=1. Both will generate same results.


TO DETERMINE P AND Q we must use the following:

1.Autocorrelation Function (ACF): It is a measure of the correlation between the the TS with a lagged version of itself. For instance at lag 5, ACF would compare series at time instant ‘t1’…’t2’ with series at instant ‘t1-5’…’t2-5’ (t1-5 and t2 being end points).

2.Partial Autocorrelation Function (PACF): This measures the correlation between the TS with a lagged version of itself but after eliminating the variations already explained by the intervening comparisons. Eg at lag 5, it will check the correlation but remove the effects already explained by lags 1 to 4.




```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
```

```python
lag_acf = acf(ts_log_diff, nlags=20)
lag_pacf = pacf(ts_log_diff, nlags=20, method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
```

```python
#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

In this plot, the two dotted lines on either sides of 0 are the confidence interevals. These can be used to determine the ‘p’ and ‘q’ values as:

p – The lag value where the PACF chart crosses the upper confidence interval for the first time. If you notice closely, in this case p=2.
q – The lag value where the ACF chart crosses the upper confidence interval for the first time. If you notice closely, in this case q=2.

So to select P and Q, it's the first time they cross a confidence level. 


Now, lets make 3 different ARIMA models considering individual as well as combined effects. I will also print the RSS for each. Please note that here RSS is for the values of residuals and not actual series.

```python
from statsmodels.tsa.arima_model import ARIMA
```

AR MODEL

```python
model = ARIMA(ts_log, order=(2, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-ts_log_diff)**2))
```

MA MODEL

```python
model = ARIMA(ts_log, order=(0, 1, 2))  
results_MA = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-ts_log_diff)**2))
```

Combined Model

```python
model = ARIMA(ts_log, order=(2, 1, 2))  
results_ARIMA = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ts_log_diff)**2))
```

NOTICE THE RSS, when combined it's much better. So basically we've got a way to predict and it's not too far off from reality.  

NOW TO TAKE IT BACK TO THE ORIGINAL SCALE

```python
#first step is to store the values into a series
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
print (predictions_ARIMA_diff.head())
```

Notice that they start from february and not January? This is because we took a lag by 1. 

To convert the DIFFERENCING to log scale, you add these differences consecutively to the base number. v

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print (predictions_ARIMA_diff_cumsum.head())
```

Next we’ve to add them to base number. For this lets create a series with all values as base number and add the differences to it.

Remember we've difference everything initially, so as is we don't know what the original number really looked like. Just fetch the first number from your original series as an example. 

```python
predictions_ARIMA_log = pd.Series(ts_log.ix[0], index=ts_log.index)
print("the original base number")
print(predictions_ARIMA_log.head())

predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
```

Last step is to take the exponent and compare with the original series.

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(ts)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-ts)**2)/len(ts)))
```

As you can see there's still some work that needs to be done. The red line is your prediction.
