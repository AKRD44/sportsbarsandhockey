```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
from myscripts import create_stats, test_stationarity,prep_hockey_data,rmse_cv
plt.rcParams['figure.figsize'] = (10, 6)
%matplotlib inline
```

```python
data=pd.read_csv("sportscenedata/sportscene.csv",index_col="date",parse_dates=True)
```

```python
data.columns=["quarter","half","year"]
data.sort_index(inplace=True)

data=data.groupby(data.index).first()
```

```python
data=data.iloc[:,[0,2]]
data.iloc[:,0]=data.iloc[:,0].str.replace(",","")

data=data.apply(pd.to_numeric)
data_end_of_month=data.fillna(0)
data_end_of_month["merged"]=data_end_of_month.iloc[:,0]+data_end_of_month.iloc[:,1]
```

```python

yearly_data=data_end_of_month.merged[data_end_of_month.index.month==8]

quarters_data=data_end_of_month.quarter[data_end_of_month.index.month!=8]

if quarters_data.shape[0]%3!=0:
    print()
    quarters_data=quarters_data.iloc[:-(quarters_data.shape[0]%3)] #not done with current year,


grouped_up=np.split(quarters_data, quarters_data.shape[0]/3)

totals=[]
for tmp in grouped_up:
    totals.append(tmp.sum())
    
```

```python
data_end_of_month.merged[data_end_of_month.index.month==8]=data_end_of_month.merged[data_end_of_month.index.month==8].values-totals
```

```python
data_end_of_month=data_end_of_month.merged.resample("M").sum()
data_end_of_month.dropna(inplace=True)

all_dates_inplay=data_end_of_month.index.date
```

```python

data_end_of_month[data_end_of_month.values<0]
```

```python
data_end_of_month=np.log(data_end_of_month)
data_end_of_month.plot()

```

```python
test_stationarity(data_end_of_month)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(data_end_of_month)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(data_end_of_month, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()


residual.dropna(inplace=True)

trend.dropna(inplace=True)
```

```python
test_stationarity(residual)
```

```python
from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(residual, nlags=20)

lag_pacf = pacf(residual, nlags=20, method='ols')
```

```python
plt.rcParams['figure.figsize'] = (15, 6)
#Plot ACF: #point where the line crosses the confidence interval for the first time is p
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(residual)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(residual)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

#Plot PACF: #point where the line crosses the confidence interval for the first time is 1
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(residual)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(residual)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA
#basically showing here that there's no autocorrelation.
```

```python
model = ARIMA(data_end_of_month, order=(0, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(residual)
plt.plot(results_AR.fittedvalues, color='red')

difference=(results_AR.fittedvalues-residual).dropna()
plt.title('RSS: %.4f'% sum(difference**2))
```

```python
hockey_data=pd.read_csv('mtl_hockey_granular.csv',usecols=["DATE","AWAY","HOME"],index_col="DATE",parse_dates=True)
```

```python
hockey_data=prep_hockey_data(hockey_data)
```

```python
from datetime import date
from dateutil.relativedelta import relativedelta
dates_in_common=np.intersect1d(hockey_data.index.date, all_dates_inplay)   
```

```python

dates_in_common=list(dates_in_common)
hockey_summed_up = pd.DataFrame(columns=hockey_data.columns)

for each_date_in_common in dates_in_common:
    dates_to_sum=[each_date_in_common]
    dates_to_sum=[date(each_date_in_common.year,each_date_in_common.month,1)-relativedelta(days=1)]+dates_to_sum
    previous_month=each_date_in_common-relativedelta(months=1)
    dates_to_sum=[date(previous_month.year,previous_month.month,1)-relativedelta(days=1)]+dates_to_sum
    #print(dates_to_sum)
    dates_to_sum=np.intersect1d(hockey_data.index.date, dates_to_sum)#can't just get the hockey data by months because sometimes there is no hockey.    
    #keep in mind that some of these dates might not exist for the business side as well,
    
    #print(dates_to_sum)
    
    #print(dates_to_sum)
    dates_to_sum_timestamp=[]
    for each in dates_to_sum:
        dates_to_sum_timestamp.append(pd.Timestamp(each))
    
    hockey_data_min_sum=pd.DataFrame(hockey_data.loc[dates_to_sum_timestamp,:].sum())
    hockey_data_min_sum=hockey_data_min_sum.T
    hockey_data_min_sum["date"]=pd.Timestamp(each_date_in_common) #error would have slipped in here
    hockey_data_min_sum=hockey_data_min_sum.set_index("date")

    if hockey_summed_up.shape[0]==0:
        hockey_summed_up=hockey_data_min_sum
    else:
        hockey_summed_up=pd.concat([hockey_summed_up,hockey_data_min_sum])

```

```python
#shapes are still not the same...
for each in data_end_of_month.index.values: #equivalent to dates_available
    #the dates come from dates_available
    #those are the ones every 3 months, of those you find what they have in common, that's dates_in_common 
    #then you then try to go back two more months, and get the three data's worth of hockey.
    #you check to make sure those 3 months of hockey exist with also np.intersection hockey_data
    #at this point I know there still are some dates that aren't part of dates_available, but that's ok
    #I sum up the 2-3 months into one entry, eventually hope to just get the 4 quarters dates that are typical
    
    #so what this is saying here is that there was business going on at those dates, but there might've been no hockey!
    #didn't take into account the times where there was no hockey. 
    if each not in hockey_summed_up.index.values:
        print(each)

```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.model_selection import cross_val_score

x_cols=[]
y_cols=[]
for each_col in merged_df.columns:
    if "merged" not in each_col:
        x_cols.append(each_col)
    else:
        y_cols.append(each_col)
        
```

```python
x_cols
```

```python

X_train=merged_df[x_cols]
y_train=merged_df.mergedRollingMean

features=X_train.columns

def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, X_train, y_train, scoring="neg_mean_squared_error", cv = 5))
	#with scoring being blank, by default this would've outputted the accuracy, ex: 95%
	#with scoring="neg_mean_squared_error", we get accuracy -1, so shows by how much you were off and it's negative
	#then with the - in front, gives you the error, but positive. 
    return(rmse)

#ELASTIC NET METhOD
#basically combines lasso and ridge. Should usually give the best result

#this picks out the best alpha for us
model_elastic = ElasticNetCV(alphas = [1, 0.1, 0.001, 0.0005]).fit(X_train, y_train) #it will estimate the best alpha based on its tests with the alphas you give it

print("lowest error being %f percent"%rmse_cv(model_elastic).mean()) #we want to see the MEAN SQUARED ERROR that the best alpha provides


```

```python
import matplotlib
#FEATURE SELECTION
#getting an idea of what's important and what isn't.  
coef = pd.Series(model_elastic.coef_, index = features)

print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")

matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)

plt.title("Coefficients in the Lasso Model")
coef.abs().sort_values().plot(kind = "barh")

```
